```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#make data for classification problem (binary classification) 
# 1000 data points
import sklearn.datasets as skd 
X, y = skd.make_moons(n_samples=1000, noise=0.178, random_state=0)
#visualize the data
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
tf.random.set_seed(42)

# make a model
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(10, activation="relu"),
        tf.keras.layers.Dense(5, activation="relu"),
        tf.keras.layers.Dense(1, activation="sigmoid"),
    ]
)

# compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=["accuracy"],
)

# train the model
history=model.fit(X, y, epochs=50, batch_size=32, verbose=1)
```

```python

# evaluate the model
model.evaluate(X, y)
```

```python
# plot the learning curve
pd.DataFrame(history.history).plot()
plt.show()
```

```python
#test the model
# test data points
X_test, y_test = skd.make_moons(n_samples=1000, noise=0.2, random_state=42)

#test the model
model.evaluate(X_test, y_test)
```
